```python
from langchain_core.prompts import PromptTemplate
from langchain_ollama import OllamaLLM
```

```python
llm = OllamaLLM(model="llama2")
```

```python
prompt = PromptTemplate(
    input_variables=["features"],
    template="""
    A vehicle insurance claim was flagged as potential fraud.

    Key contributing factors:
    {features}

    Explain in clear business terms why this claim may be fraudulent.
    """
)

chain = prompt | llm

response = chain.invoke({
    "features": "insured_hobbies, incident_severity"
})

print(response)
```
