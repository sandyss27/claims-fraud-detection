```python
import pandas as pd
import sqlite3
import numpy as np
```

```python
df = pd.read_csv("dataset/insurance_claims.csv")
df.head()
```

```python
conn = sqlite3.connect("claims.db")
df.to_sql("claims", conn, if_exists="replace", index=False)
conn.close()
```

```python
conn = sqlite3.connect("claims.db")
query = "PRAGMA table_info(claims);"
schema = pd.read_sql_query(query, conn)
conn.close()
schema
```
